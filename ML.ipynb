```python
def remove_outliers_by_quantile(df, features, lower=0.02, upper=0.98):
    df_filtered = df.copy()
    for col in features:
        q_low = df[col].quantile(lower)
        q_high = df[col].quantile(upper)
        df_filtered = df_filtered[(df_filtered[col] >= q_low) & (df_filtered[col] <= q_high)]
    return df_filtered

# Áp dụng
features = [
    'current_ratio', 'quick_ratio', 'cash_ratio',
    'debt_to_assets', 'debt_to_equity',
    'receivables_turnover', 'inventory_turnover',
    'ROA', 'ROE', 'net_profit_margin', 'operating_cashflow_ratio']
df = remove_outliers_by_quantile(df, features, 0.02, 0.98)
```

  Chọn eps, min_samples, K cho model DBSCAN+KMEAN

```python
from sklearn.cluster import DBSCAN, KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

# Các giá trị thử
eps_list = [1.0, 1.5, 2.0, 2.5]
min_samples_list = [3, 5, 8]
k_list = [3, 4, 5, 6]

results = []

# Vòng lặp thử nhiều tổ hợp
for eps in eps_list:
    for min_samples in min_samples_list:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        db_labels = dbscan.fit_predict(X_scaled)

        # Lọc điểm core (loại outlier)
        mask_core = db_labels != -1
        X_core = X_scaled[mask_core]

        # Bỏ qua nếu dữ liệu sau DBSCAN quá ít
        if len(X_core) < 50:
            continue

        for k in k_list:
            kmeans = KMeans(n_clusters=k, init='k-means++', n_init=20, random_state=42)
            try:
                kmeans_labels = kmeans.fit_predict(X_core)
                score = silhouette_score(X_core, kmeans_labels)
                results.append({
                    'eps': eps,
                    'min_samples': min_samples,
                    'k': k,
                    'silhouette_score': round(score, 4),
                    'n_samples': len(X_core)
                })
            except:
                continue

# Chuyển thành DataFrame và in top 5
results_df = pd.DataFrame(results)
top_5 = results_df.sort_values(by='silhouette_score', ascending=False).head(5)
print("Top 5 tổ hợp có Silhouette Score cao nhất:")
print(top_5.to_string(index=False))
```

```python
import hdbscan
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Chạy HDBSCAN
clusterer = hdbscan.HDBSCAN(min_cluster_size=15, min_samples=5, metric='euclidean')
df['cluster'] = clusterer.fit_predict(df[features])

# 2. Map cụm thành nhãn
df['cluster'] = df['cluster'].astype(str)
cluster_labels = {
    '-1': '-1',
    '0': '0',
    '1': '1',
    '2': '2'
}
df['label'] = df['cluster'].map(cluster_labels).fillna('Cụm khác')

# 3. PCA để trực quan hóa
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df[features])
df['pca1'] = pca_result[:, 0]
df['pca2'] = pca_result[:, 1]

# 4. Vẽ phân cụm
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='pca1', y='pca2', hue='label', palette='Set2', s=50)
plt.title('Phân cụm sức khỏe tài chính bằng HDBSCAN', fontsize=14)
plt.legend(title='Cụm')
plt.grid(True)
plt.show()

# 5. In số điểm mỗi cụm
print(" Số điểm mỗi cụm:")
print(df['label'].value_counts())

# 6. Tính Silhouette Score (loại cụm -1 vì là outlier)
valid_mask = df['cluster'] != '-1'
score = silhouette_score(df.loc[valid_mask, features], df.loc[valid_mask, 'cluster'])
print(f"\n Silhouette Score (không tính cụm Cảnh báo): {score:.4f}")
```
